# 02. Decorators, Generators, and Context Managers

These are powerful Python features used in real-world applications.

What this topic is:
Decorators change the behavior of functions, generators yield values one at a time, and context managers handle setup and cleanup neatly.

Where we use it:
They are commonly used in web frameworks, logging, database connections, file handling, and streaming data.

Why we use it:
They reduce repetition, improve resource management, and keep code cleaner and safer.

```python
# Decorator example
def uppercase(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper

@uppercase
def greet():
    return 'hello'

print(greet())
```

```python
# Generator example
def count_down(n):
    while n > 0:
        yield n
        n -= 1

for value in count_down(3):
    print(value)
```

```python
# Context manager example
from contextlib import contextmanager

@contextmanager
def open_file(name):
    f = open(name, 'w', encoding='utf-8')
    try:
        yield f
    finally:
        f.close()

with open_file('advanced_example.txt') as f:
    f.write('Advanced Python')
```
