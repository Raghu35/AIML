# 01. Advanced Python Basics

This notebook revisits core concepts with a more advanced perspective.

What this topic is:
Advanced Python basics include list comprehensions, lambda functions, and functional-style programming patterns.

Where we use it:
These techniques are used in data cleaning, short scripts, and compact code for processing collections.

Why we use it:
They make code shorter, more readable, and often more expressive for common tasks.

```python
# List comprehension
squares = [x * x for x in range(10)]
print(squares)
```

```python
# Lambda function
double = lambda x: x * 2
print(double(7))
```

```python
# Map, filter, reduce style example
numbers = [1, 2, 3, 4, 5]
print(list(map(lambda x: x * 2, numbers)))
print(list(filter(lambda x: x % 2 == 0, numbers)))
```
