# 04. Multithreading and Asyncio

This notebook shows how to run tasks concurrently in Python.

What this topic is:
Multithreading and asyncio allow Python programs to handle multiple tasks efficiently.

Where we use it:
They are used in web servers, network requests, web scraping, and other input/output-heavy applications.

Why we use it:
They improve performance by avoiding unnecessary waiting during slow operations.

```python
# Threading example
import threading
import time

def task():
    time.sleep(1)
    print('Task done')

t = threading.Thread(target=task)
t.start()
t.join()
```

```python
# Asyncio example
import asyncio

async def main():
    print('Start')
    await asyncio.sleep(0.2)
    print('End')

asyncio.run(main())
```
