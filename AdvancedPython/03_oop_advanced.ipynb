# 03. Advanced OOP

This notebook shows inheritance, polymorphism, and encapsulation.

What this topic is:
Advanced OOP focuses on organizing code into reusable classes and controlling how objects behave.

Where we use it:
It is used in application design, APIs, frameworks, and large software systems.

Why we use it:
It makes code modular, reusable, and easier to maintain as projects grow.

```python
# Encapsulation example
class BankAccount:
    def __init__(self, balance):
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount

    def get_balance(self):
        return self._balance

account = BankAccount(100)
account.deposit(50)
print(account.get_balance())
```

```python
# Inheritance and polymorphism
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

shapes = [Rectangle(4, 5)]
for shape in shapes:
    print(shape.area())
```
