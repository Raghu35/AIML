# 05. Package-Based Examples

This notebook contains common useful packages for Python development.

What this topic is:
Packages such as NumPy, Pandas, and Matplotlib provide ready-made tools for data work and visualization.

Where we use it:
They are used in data analysis, machine learning, scientific computing, dashboards, and reporting.

Why we use it:
They save time, improve reliability, and let us solve complex tasks without building everything from scratch.

```python
# NumPy example
import numpy as np
arr = np.array([1, 2, 3, 4])
print(arr * 2)
```

```python
# Pandas example
import pandas as pd
df = pd.DataFrame({'Name': ['Asha', 'Bob'], 'Age': [25, 30]})
print(df)
```

```python
# Matplotlib example
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [3, 2, 1])
plt.title('Simple Plot')
plt.show()
```
