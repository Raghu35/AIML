# Module 3 — Dataclasses

**What it is:** `@dataclass` auto-generates `__init__`, `__repr__`, and comparison methods for classes that primarily store data. Fields can have defaults, type hints, and frozen (immutable) mode.

**Where we use it:** DTOs, configuration objects, API request/response models, and any plain data container before reaching for heavier ORM or validation libraries.

**Why we use it:** Dataclasses reduce boilerplate while keeping code readable and type-friendly.

```python
from dataclasses import dataclass, field, asdict

@dataclass
class Student:
    name: str
    age: int
    grade: str = "B"
    hobbies: list[str] = field(default_factory=list)

s1 = Student("Asha", 21, "A", ["Python", "Chess"])
s2 = Student("Riya", 20)
print(s1)
print("As dict:", asdict(s1))

# frozen=True makes instances immutable (like a named tuple with methods)
@dataclass(frozen=True)
class Point:
    x: float
    y: float

p = Point(3.0, 4.0)
print("Point:", p)
```

## Post-init and ordering

```python
from dataclasses import dataclass, field

@dataclass(order=True)
class Task:
    priority: int
    title: str = field(compare=False)

    def __post_init__(self):
        if self.priority < 1:
            raise ValueError("Priority must be >= 1")

tasks = [Task(2, "Write tests"), Task(1, "Fix bug"), Task(3, "Deploy")]
print("Sorted by priority:", sorted(tasks))
```
