# Module 3 — Logging

**What it is:** The `logging` module records events with severity levels (DEBUG, INFO, WARNING, ERROR, CRITICAL), formatters, and handlers that write to console, files, or remote systems.

**Where we use it:** Production applications, debugging pipelines, audit trails, and monitoring long-running services.

**Why we use it:** Structured logs are searchable, persistent, and configurable — far more maintainable than scattered `print()` statements.

```python
import logging
from pathlib import Path

OUTPUT = Path("Module3/output")
OUTPUT.mkdir(parents=True, exist_ok=True)
log_file = OUTPUT / "app.log"

# Basic configuration: level, format, handlers
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s | %(levelname)s | %(name)s | %(message)s",
    handlers=[
        logging.StreamHandler(),
        logging.FileHandler(log_file, encoding="utf-8"),
    ],
    force=True,
)

logger = logging.getLogger("aiml.demo")

logger.debug("Detailed diagnostic info")
logger.info("Application started")
logger.warning("Disk space low")
logger.error("Failed to connect to database")

print("\nLog file contents:")
print(log_file.read_text(encoding="utf-8"))
```

## Rotating file handler

```python
import logging
from logging.handlers import RotatingFileHandler
from pathlib import Path

OUTPUT = Path("Module3/output")
rotate_log = OUTPUT / "rotate.log"

logger = logging.getLogger("rotate.demo")
logger.setLevel(logging.INFO)
logger.handlers.clear()

handler = RotatingFileHandler(rotate_log, maxBytes=100, backupCount=2, encoding="utf-8")
handler.setFormatter(logging.Formatter("%(levelname)s: %(message)s"))
logger.addHandler(handler)

for i in range(5):
    logger.info(f"Log entry number {i}")

print("Rotating log size (bytes):", rotate_log.stat().st_size)
```
