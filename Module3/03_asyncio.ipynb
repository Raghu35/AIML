# Module 3 — AsyncIO

**What it is:** AsyncIO provides cooperative concurrency using `async`/`await`. A single thread runs many coroutines, switching between them when they await I/O instead of blocking.

**Where we use it:** High-concurrency web servers, chat applications, websockets, and thousands of simultaneous network requests.

**Why we use it:** It scales I/O-bound workloads with lower memory overhead than one thread per connection.

```python
import asyncio

async def greet(name, delay):
    print(f"Hello {name}, waiting {delay}s")
    await asyncio.sleep(delay)
    print(f"Goodbye {name}")

async def main():
    await asyncio.gather(
        greet("Asha", 0.3),
        greet("Riya", 0.1),
        greet("Bob", 0.2),
    )

asyncio.run(main())
```

## Tasks and timeouts

```python
import asyncio

async def slow_fetch(item_id):
    await asyncio.sleep(0.5)
    return f"data-{item_id}"

async def main():
    task = asyncio.create_task(slow_fetch(42))
    try:
        result = await asyncio.wait_for(task, timeout=1.0)
        print("Result:", result)
    except asyncio.TimeoutError:
        print("Request timed out")

asyncio.run(main())
```

## Async context manager pattern

```python
import asyncio
from contextlib import asynccontextmanager

@asynccontextmanager
async def connection():
    print("Opening connection")
    try:
        yield "conn-123"
    finally:
        print("Closing connection")

async def main():
    async with connection() as conn:
        print("Using", conn)
        await asyncio.sleep(0.1)

asyncio.run(main())
```
