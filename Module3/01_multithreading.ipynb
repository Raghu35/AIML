# Module 3 — Multithreading

**What it is:** Multithreading runs multiple threads within one process, sharing the same memory space. Python's `threading` module is best for I/O-bound tasks (network, disk) where threads wait on external resources.

**Where we use it:** Concurrent downloads, web scraping, GUI responsiveness, and parallel I/O operations.

**Why we use it:** While waiting on one task, other threads can progress — improving throughput for I/O-heavy workloads.

```python
import threading
import time

def download_simulation(name, seconds):
    print(f"[{name}] starting")
    time.sleep(seconds)
    print(f"[{name}] finished")

# Create and start threads
threads = [
    threading.Thread(target=download_simulation, args=("Task-A", 0.5)),
    threading.Thread(target=download_simulation, args=("Task-B", 0.3)),
]
for t in threads:
    t.start()
for t in threads:
    t.join()  # wait for all threads to complete

print("All threads done")
```

## Thread locks for shared state

```python
import threading

counter = 0
lock = threading.Lock()

def increment(times):
    global counter
    for _ in range(times):
        with lock:  # only one thread enters at a time
            counter += 1

workers = [threading.Thread(target=increment, args=(100_000,)) for _ in range(4)]
for w in workers:
    w.start()
for w in workers:
    w.join()

print("Final counter:", counter)
```

## ThreadPoolExecutor

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def fetch(url_id):
    time.sleep(0.2)
    return f"Response from {url_id}"

urls = ["api/users", "api/orders", "api/products"]

with ThreadPoolExecutor(max_workers=3) as executor:
    futures = {executor.submit(fetch, u): u for u in urls}
    for future in as_completed(futures):
        print(future.result())
```
