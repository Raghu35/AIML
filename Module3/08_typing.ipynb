# Module 3 — Typing

**What it is:** The `typing` module adds optional static type hints — `list[int]`, `Optional[str]`, `Union`, `Callable`, `Generic`, and `Protocol` — that tools like mypy and IDEs use for analysis.

**Where we use it:** Large codebases, public libraries, team projects, and anywhere early bug detection and self-documenting signatures matter.

**Why we use it:** Type hints improve readability and catch category errors before runtime without changing Python's dynamic nature.

```python
from typing import Optional, Union, Callable, TypeVar, Generic

# Basic hints on parameters and return values
def greet(name: str, times: int = 1) -> str:
    return (f"Hello {name}! " * times).strip()

# Optional — value or None
def find_user(user_id: int) -> Optional[dict]:
    database = {1: {"name": "Asha"}}
    return database.get(user_id)

# Union — one of several types
def parse_id(raw: Union[str, int]) -> int:
    return int(raw)

# Callable — function type
def apply_twice(func: Callable[[int], int], value: int) -> int:
    return func(func(value))

print(greet("Asha", 2))
print(find_user(1))
print(parse_id("42"))
print(apply_twice(lambda x: x + 1, 5))
```

## Generics and TypedDict

```python
from typing import Generic, TypeVar, TypedDict

T = TypeVar("T")

class Stack(Generic[T]):
    def __init__(self) -> None:
        self._items: list[T] = []

    def push(self, item: T) -> None:
        self._items.append(item)

    def pop(self) -> T:
        return self._items.pop()

class UserRecord(TypedDict):
    name: str
    age: int

int_stack = Stack[int]()
int_stack.push(10)
int_stack.push(20)
print("Stack pop:", int_stack.pop())

user: UserRecord = {"name": "Asha", "age": 21}
print("TypedDict:", user)
```
