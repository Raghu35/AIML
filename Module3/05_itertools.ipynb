# Module 3 — Itertools

**What it is:** `itertools` offers iterator-building blocks for infinite sequences, combinations, permutations, grouping, chaining, and slicing — all without storing full results in memory.

**Where we use it:** Combinatorial problems, pagination, streaming pipelines, cartesian products in testing, and efficient data transformations.

**Why we use it:** It composes lazy iterators that are memory-efficient and expressive for sequence manipulation.

```python
import itertools as it

# Infinite iterators — use islice to bound them
print("First 5 count:", list(it.islice(it.count(10, 2), 5)))
print("Cycle ABC x4:", list(it.islice(it.cycle("ABC"), 7)))

# Combinatorics
print("Permutations AB:", list(it.permutations("AB", 2)))
print("Combinations 123:", list(it.combinations([1, 2, 3], 2)))
print("Product:", list(it.product([1, 2], ["x", "y"])))

# Group consecutive items (input must be sorted by key for correct groups)
records = [("A", 90), ("A", 85), ("B", 80), ("B", 78)]
for key, group in it.groupby(records, key=lambda r: r[0]):
    print(f"Group {key}:", list(group))

# Chain and zip
print("Chain:", list(it.chain([1, 2], [3, 4])))
print("Zip longest:", list(it.zip_longest([1, 2, 3], ["a", "b"], fillvalue="-")))
```
