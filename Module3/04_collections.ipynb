# Module 3 — Collections

**What it is:** The `collections` module provides specialized container types — `Counter`, `defaultdict`, `deque`, `namedtuple`, and `OrderedDict` — that extend built-in dict and list behavior.

**Where we use it:** Frequency counting, grouping data, efficient queue operations, lightweight records, and structured intermediate results in ETL pipelines.

**Why we use it:** These types express common patterns more clearly and often perform better than manual dict/list workarounds.

```python
from collections import Counter, defaultdict, deque, namedtuple, OrderedDict

# Counter — count hashable items
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
counts = Counter(words)
print("Most common:", counts.most_common(2))
print("Apple count:", counts["apple"])

# defaultdict — auto-create missing keys
sales = defaultdict(int)
for product in ["pen", "book", "pen", "pen"]:
    sales[product] += 1
print("Sales:", dict(sales))

# deque — fast append/pop from both ends
queue = deque(["task1", "task2"])
queue.append("task3")
queue.popleft()
print("Queue:", list(queue))

# namedtuple — lightweight immutable record
Point = namedtuple("Point", ["x", "y"])
origin = Point(0, 0)
print("Point:", origin, origin.x)

# OrderedDict — remembers insertion order (dict does too since 3.7, but API differs)
registry = OrderedDict()
registry["first"] = 1
registry["second"] = 2
print("Ordered keys:", list(registry.keys()))
```
