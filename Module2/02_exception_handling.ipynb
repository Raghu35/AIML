# Module 2 — Exception Handling

**What it is:** Exceptions are runtime errors that Python raises when something goes wrong. Exception handling lets you catch, respond to, or re-raise those errors gracefully.

**Where we use it:** Input validation, API calls, file operations, database work, and anywhere failure is possible and must not crash the whole program.

**Why we use it:** It keeps applications stable, provides meaningful error messages, and separates normal logic from error-recovery logic.

## try / except / else / finally

```python
# try: code that might fail
# except: handle specific errors
# else: runs only if no exception occurred
# finally: always runs (cleanup)

try:
    result = 10 / 2
    print("Result:", result)
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division succeeded with no errors")
finally:
    print("Cleanup block always executes")
```

## Catching multiple exceptions and raising custom errors

```python
# Catch different exception types separately.
# raise lets you signal problems explicitly.

class AgeTooSmallError(Exception):
    'Raised when age is below the allowed minimum.'

def validate_age(raw_value):
    try:
        age = int(raw_value)
    except ValueError as exc:
        raise ValueError("Age must be a number") from exc

    if age < 18:
        raise AgeTooSmallError(f"Age {age} is too small")
    return age

for test_value in ["25", "abc", "15"]:
    try:
        valid_age = validate_age(test_value)
        print(f"{test_value!r} -> valid: {valid_age}")
    except ValueError as exc:
        print(f"{test_value!r} -> ValueError: {exc}")
    except AgeTooSmallError as exc:
        print(f"{test_value!r} -> Custom error: {exc}")
```

## Exception chaining and best practices

```python
# Use specific exceptions, avoid bare except, and preserve context with 'from'.

def read_positive_number(text):
    try:
        value = int(text)
    except ValueError as exc:
        raise ValueError(f"Invalid number: {text!r}") from exc

    if value <= 0:
        raise ValueError("Number must be positive")
    return value

try:
    read_positive_number("zero")
except ValueError as exc:
    print("Error:", exc)
    print("Caused by:", exc.__cause__)
```
