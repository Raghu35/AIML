# Module 2 — Decorators

**What it is:** A decorator is a function that wraps another function (or class) to extend or modify its behavior without changing its source code. The `@decorator` syntax is syntactic sugar for `func = decorator(func)`.

**Where we use it:** Web frameworks (route handlers), logging, authentication, caching, timing, and validation in production code.

**Why we use it:** Decorators keep cross-cutting concerns separate from business logic and reduce repetitive boilerplate.

## Function decorator basics

```python
# A decorator receives a function and returns a new wrapper function.

def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper

@uppercase_decorator
def greet(name):
    return f"hello {name}"

print(greet("Asha"))
```

## Decorators with arguments using `functools.wraps`

```python
from functools import wraps
import time

# Preserve original function metadata (__name__, __doc__) with @wraps.

def repeat(times):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(times):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

@repeat(2)
@timer
def compute():
    'Simulate a small calculation.'
    return sum(range(100_000))

print("Result:", compute())
print("Function name preserved:", compute.__name__)
```

## Class-based decorator

```python
# Classes with __call__ can act as decorators too.

class CallCounter:
    def __init__(self, func):
        self.func = func
        self.calls = 0

    def __call__(self, *args, **kwargs):
        self.calls += 1
        print(f"Call #{self.calls}")
        return self.func(*args, **kwargs)

@CallCounter
def add(a, b):
    return a + b

print(add(2, 3))
print(add(10, 20))
```
