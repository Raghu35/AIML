# Module 2 — File Handling

**What it is:** File handling reads from and writes to files on disk using paths, modes (`r`, `w`, `a`), and encodings. `pathlib.Path` offers a modern, object-oriented way to work with file paths.

**Where we use it:** Logs, configuration files, CSV/JSON datasets, reports, and persisting application data between runs.

**Why we use it:** Most real programs need durable storage outside memory; proper file handling makes data portable and reproducible.

## Text files with `open()` and `pathlib`

```python
from pathlib import Path

# Use a dedicated output folder so notebooks stay tidy.
OUTPUT = Path("Module2/output")
OUTPUT.mkdir(parents=True, exist_ok=True)

text_path = OUTPUT / "sample.txt"

# Write text (w = write, overwrites existing content)
text_path.write_text("Hello from Python\nLine 2\n", encoding="utf-8")

# Read entire file
content = text_path.read_text(encoding="utf-8")
print("File content:")
print(content)

# Append mode adds without deleting existing data
with text_path.open("a", encoding="utf-8") as handle:
    handle.write("Appended line\n")

print("After append:", text_path.read_text(encoding="utf-8"))
```

## CSV, JSON, and XML

```python
import csv
import json
import xml.etree.ElementTree as ET
from pathlib import Path

OUTPUT = Path("Module2/output")

# CSV — tabular data
csv_path = OUTPUT / "students.csv"
with csv_path.open("w", newline="", encoding="utf-8") as handle:
    writer = csv.DictWriter(handle, fieldnames=["name", "age", "grade"])
    writer.writeheader()
    writer.writerow({"name": "Asha", "age": 21, "grade": "A"})
    writer.writerow({"name": "Riya", "age": 20, "grade": "B"})
print("CSV written:", csv_path)

# JSON — structured data interchange
json_path = OUTPUT / "student.json"
with json_path.open("w", encoding="utf-8") as handle:
    json.dump({"name": "Asha", "skills": ["Python", "ML"]}, handle, indent=2)
print("JSON:", json_path.read_text(encoding="utf-8"))

# XML — hierarchical markup
xml_path = OUTPUT / "student.xml"
root = ET.Element("student")
ET.SubElement(root, "name").text = "Asha"
ET.SubElement(root, "age").text = "21"
ET.ElementTree(root).write(xml_path, encoding="utf-8", xml_declaration=True)
print("XML written:", xml_path)
```

## YAML (optional) and reading line-by-line

```python
from pathlib import Path

OUTPUT = Path("Module2/output")

# Read large files line-by-line to avoid loading everything into memory.
text_path = OUTPUT / "sample.txt"
print("Lines:")
for line in text_path.read_text(encoding="utf-8").splitlines():
    print(" ", line)

try:
    import yaml
except ImportError:
    print("PyYAML not installed. Run: pip install pyyaml")
else:
    yaml_path = OUTPUT / "config.yaml"
    data = {"app": "AIML", "debug": True, "tags": ["python", "ml"]}
    yaml_path.write_text(yaml.dump(data, sort_keys=False), encoding="utf-8")
    print("YAML:", yaml.safe_load(yaml_path.read_text(encoding="utf-8")))
```
