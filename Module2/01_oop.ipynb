# Module 2 — Object-Oriented Programming (OOP)

**What it is:** OOP organizes code into *classes* (blueprints) and *objects* (instances) that bundle data and behavior together.

**Where we use it:** Application design, APIs, game engines, and any system where you model real-world entities like users, orders, or bank accounts.

**Why we use it:** It improves reuse, readability, and scalability by grouping related logic and protecting internal state.

## Classes, objects, and constructors

```python
# A class defines attributes (data) and methods (behavior).
# __init__ is the constructor — it runs when a new object is created.

class Dog:
    def __init__(self, name, breed):
        self.name = name      # instance attribute
        self.breed = breed

    def bark(self):
        return f"{self.name} says woof!"

# An object is a concrete instance of the class.
my_dog = Dog("Bruno", "Labrador")
print(my_dog.bark())
print(f"Name: {my_dog.name}, Breed: {my_dog.breed}")
```

## Inheritance and polymorphism

```python
# Inheritance lets a child class reuse and extend a parent class.
# Polymorphism means different classes can respond to the same method name differently.

class Animal:
    def speak(self):
        return "Some sound"

class Cat(Animal):
    def speak(self):
        return "Meow"

class Dog(Animal):
    def speak(self):
        return "Woof"

# Polymorphism: one interface, many implementations.
animals = [Cat(), Dog(), Animal()]
for animal in animals:
    print(animal.speak())
```

## Encapsulation and abstraction

```python
# Encapsulation hides internal details using naming conventions and controlled access.
# Abstraction defines *what* an object must do without specifying *how*.

from abc import ABC, abstractmethod

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # name mangling discourages direct access

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

account = BankAccount("Asha", 100)
account.deposit(50)
print("Balance:", account.get_balance())

rect = Rectangle(4, 5)
print("Area:", rect.area())
```

## Magic methods (`__str__`, `__repr__`)

```python
# Magic methods customize how objects behave with built-in operations.

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} (${self.price})"

    def __repr__(self):
        return f"Product(name={self.name!r}, price={self.price})"

item = Product("Notebook", 12.99)
print(str(item))   # user-friendly
print(repr(item))  # developer-friendly
```
