# 07. Collections and Regex

This notebook demonstrates useful collections and regular expressions.

What this topic is:
Collections provide specialized data containers, and regular expressions help match and search text patterns.

Where we use it:
They are used in text processing, validation, parsing, and data extraction tasks.

Why we use it:
They make it easier to process and validate large amounts of text efficiently.

```python
# Counter example
from collections import Counter
words = ['apple', 'banana', 'apple', 'orange']
print(Counter(words))
```

```python
# Regex example
import re
text = 'Python123 is great'
matches = re.findall(r'+', text)
print(matches)
```
