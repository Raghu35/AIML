# 07. Collections and Regex

This notebook demonstrates useful collections and regular expressions.

**What it is:** The `collections` module provides specialized containers like `Counter` and `defaultdict`. Regular expressions (`re`) match and extract text patterns using concise syntax.

**Where we use it:** Text processing, validation, parsing log files, counting word frequencies, and data extraction from unstructured strings.

**Why we use it:** They make it easier to process and validate large amounts of text efficiently without writing complex manual loops.

> For full Module 3 coverage see [Module3/04_collections.ipynb](../Module3/04_collections.ipynb) and [Module3/05_itertools.ipynb](../Module3/05_itertools.ipynb).

```python
# Counter — count occurrences of hashable items in an iterable
from collections import Counter, defaultdict, deque

words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
print("Counter:", Counter(words))
print("Most common 2:", Counter(words).most_common(2))

# defaultdict — automatically creates missing keys with a default factory
scores = defaultdict(list)
scores['Asha'].append(90)
scores['Asha'].append(85)
scores['Riya'].append(88)
print("defaultdict:", dict(scores))

# deque — double-ended queue with fast append/pop on both ends
queue = deque(['task1', 'task2'])
queue.append('task3')
queue.popleft()
print("deque:", list(queue))
```

```python
# Regex — find digit sequences and word patterns in text
import re

text = 'Python123 is great and has 42 features'
digits = re.findall(r'\d+', text)           # one or more digits
words = re.findall(r'[A-Za-z]+', text)      # alphabetic words
print("Digits found:", digits)
print("Words found:", words)

# Validate email-like pattern
emails = 'Contact alice@example.com or bob@test.org'
pattern = re.compile(r'\b[\w.-]+@[\w.-]+\.\w+\b')
print("Emails:", pattern.findall(emails))
```
