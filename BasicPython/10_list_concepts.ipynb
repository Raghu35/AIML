# List Concepts

This notebook demonstrates common list operations in Python.

What this topic is:
Lists are ordered containers that store multiple values and support many useful operations.

Where we use it:
They are used for storing user data, processing sequences, and handling collections of values.

Why we use it:
They are one of the most common and flexible structures in Python.

```python
# 1. Creating a list
phones = ["iPhone", "Samsung", "OnePlus", "Pixel"]
print("Original list:", phones)
```

```python
# 2. Accessing elements
print("First phone:", phones[0])
print("Last phone:", phones[-1])
```

```python
# 3. Slicing a list
print("First three phones:", phones[0:3])
```

```python
# 4. Adding elements
phones.append("Xiaomi")
print("After appending:", phones)
```

```python
# 5. Inserting an element at a specific position
phones.insert(2, "Vivo")
print("After inserting at index 2:", phones)
```

```python
# 6. Removing elements
phones.remove("Samsung")
print("After removing Samsung:", phones)
```

```python
# 7. Replacing an element
phones[1] = "Motorola"
print("After replacing index 1:", phones)
```

```python
# 8. Sorting and reversing
phones.sort()
print("Sorted list:", phones)
phones.reverse()
print("Reversed list:", phones)
```

```python
# 9. Finding length of a list
print("Number of phones:", len(phones))
```

```python
# 10. Checking membership
print("Is iPhone present?", "iPhone" in phones)
```

```python
# 11. Using list comprehension
squared_numbers = [x * x for x in range(1, 6)]
print("Squared numbers:", squared_numbers)
```

```python
# 12. Nested list example
phone_details = [["iPhone", 999], ["Samsung", 799], ["OnePlus", 699]]
print("Phone details:", phone_details)
print("Second phone price:", phone_details[1][1])
```

```python
# 13. Summary note
print("\nList summary:")
print("- Lists store multiple items in one variable")
print("- They are mutable, meaning elements can be changed")
print("- They support indexing, slicing, sorting, and more")
```
