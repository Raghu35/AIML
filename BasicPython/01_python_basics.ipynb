# 01. Python Basics

This notebook covers the first building blocks of Python.

What this topic is:
Variables, literals, operators, and basic input/output are the first building blocks of Python programming.

Where we use it:
They are used in every Python script, from automation to data analysis and web development.

Why we use it:
They help you understand how data is stored and how programs perform simple tasks.

```python
# Variables store values that we can reuse later
name = "Alice"
age = 25
print("Name:", name)
print("Age:", age)
```

```python
# Basic arithmetic operations with numbers
num1 = 10
num2 = 5
print("Sum:", num1 + num2)
print("Difference:", num1 - num2)
print("Product:", num1 * num2)
```

```python
# String formatting
message = f"Hello {name}, you are {age} years old."
print(message)
```
