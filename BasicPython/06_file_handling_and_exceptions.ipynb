# 06. File Handling and Exceptions

This notebook covers reading and writing files and handling errors.

What this topic is:
File handling allows us to read and write data from files, while exceptions help manage errors safely.

Where we use it:
They are used in data storage, logs, configuration files, and user input handling.

Why we use it:
They make programs more robust and reliable in real-world situations.

```python
# Write and read a file
with open('sample.txt', 'w', encoding='utf-8') as f:
    f.write('Hello from Python')

with open('sample.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# Exception handling
try:
    result = 10 / 0
except ZeroDivisionError:
    print('Cannot divide by zero')
finally:
    print('This always runs')
```

```python
# Exceptions, file handling, CSV, JSON, XML, and YAML
import csv
import json
import xml.etree.ElementTree as ET
from pathlib import Path

class TooSmallError(Exception):
    pass

try:
    age = int("25")
    if age < 18:
        raise TooSmallError("Age is too small")
    print("Age is valid")
except ValueError as exc:
    print("ValueError:", exc)
except TooSmallError as exc:
    print("Custom error:", exc)
else:
    print("No exception occurred")
finally:
    print("This always runs")

folder = Path("BasicPython/Level_1_Programs/output")
folder.mkdir(parents=True, exist_ok=True)

text_path = folder / "sample.txt"
text_path.write_text("Hello from Python\n", encoding="utf-8")
print("Text:", text_path.read_text(encoding="utf-8").strip())

csv_path = folder / "sample.csv"
with csv_path.open("w", newline="", encoding="utf-8") as handle:
    writer = csv.writer(handle)
    writer.writerow(["name", "age"])
    writer.writerow(["Asha", 21])

json_path = folder / "sample.json"
with json_path.open("w", encoding="utf-8") as handle:
    json.dump({"name": "Asha", "age": 21}, handle)

xml_path = folder / "sample.xml"
root = ET.Element("student")
ET.SubElement(root, "name").text = "Asha"
ET.SubElement(root, "age").text = "21"
ET.ElementTree(root).write(xml_path, encoding="utf-8", xml_declaration=True)

try:
    import yaml
except ImportError:
    print("PyYAML not installed. Install it with: pip install pyyaml")
else:
    yaml_path = folder / "sample.yaml"
    yaml.safe_dump({"name": "Asha", "skills": ["Python", "Data Science"]}, open(yaml_path, "w"), sort_keys=False)
    print("YAML created")
```
