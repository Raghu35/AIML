# 06. File Handling and Exceptions

This notebook covers reading and writing files and handling errors.

What this topic is:
File handling allows us to read and write data from files, while exceptions help manage errors safely.

Where we use it:
They are used in data storage, logs, configuration files, and user input handling.

Why we use it:
They make programs more robust and reliable in real-world situations.

```python
# Write and read a file
with open('sample.txt', 'w', encoding='utf-8') as f:
    f.write('Hello from Python')

with open('sample.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# Exception handling
try:
    result = 10 / 0
except ZeroDivisionError:
    print('Cannot divide by zero')
finally:
    print('This always runs')
```
