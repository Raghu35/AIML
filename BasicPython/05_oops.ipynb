# 05. Object-Oriented Programming

This notebook introduces classes, objects, and inheritance.

What this topic is:
OOP organizes code into classes and objects to model real-world entities and behavior.

Where we use it:
It is used in application design, frameworks, game development, and large software systems.

Why we use it:
It makes code reusable, modular, and easier to scale.

```python
# Class and object example
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        return f"{self.name} says woof!"

my_dog = Dog("Bruno")
print(my_dog.bark())
```

```python
# Inheritance example
class Animal:
    def speak(self):
        return "Animal sound"

class Cat(Animal):
    def speak(self):
        return "Meow"

print(Cat().speak())
```
