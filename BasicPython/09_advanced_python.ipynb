# 09. Advanced Python Concepts

This notebook covers decorators, generators, context managers, and more.

**What it is:** Decorators wrap functions to extend behavior; generators yield values lazily; context managers handle setup/teardown with the `with` statement.

**Where we use it:** Web frameworks, logging, streaming data, file/database connections, and production libraries.

**Why we use it:** They reduce boilerplate, manage resources safely, and keep cross-cutting logic separate from business code.

```python
# Decorator example
def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper

@uppercase_decorator
def greet():
    return 'hello world'

print(greet())
```

```python
# Generator example
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for value in countdown(3):
    print(value)
```

```python
# Context manager example
from contextlib import contextmanager

@contextmanager
def file_manager(filename):
    f = open(filename, 'w', encoding='utf-8')
    try:
        yield f
    finally:
        f.close()

with file_manager('advanced_example.txt') as f:
    f.write('Advanced Python')
```

```python
# Advanced topics: lambda, map/filter/reduce, recursion, args/kwargs, iterators, generators, decorators, class decorators, and context managers
from functools import reduce
from contextlib import contextmanager

square = lambda x: x * x
print("Lambda:", square(6))

numbers = [1, 2, 3, 4, 5]
print("Map:", list(map(lambda x: x * 2, numbers)))
print("Filter:", list(filter(lambda x: x % 2 == 0, numbers)))
print("Reduce:", reduce(lambda acc, x: acc + x, numbers, 0))


def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print("Factorial:", factorial(5))


def show_args(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

show_args(1, 2, 3, name="Asha", role="student")

class NumberIterator:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.limit:
            raise StopIteration
        self.current += 1
        return self.current

print("Iterator:", list(NumberIterator(3)))


def number_generator(limit):
    for item in range(limit):
        yield item

print("Generator:", list(number_generator(4)))

class DecoratorExample:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        print("Calling decorated function")
        return self.func(*args, **kwargs)

@DecoratorExample
def greet(name):
    return f"Hello {name}"

print(greet("Asha"))

@contextmanager
def sample_context(path):
    file_obj = open(path, "w", encoding="utf-8")
    try:
        yield file_obj
    finally:
        file_obj.close()

with sample_context("BasicPython/Level_1_Programs/output/advanced_context.txt") as handle:
    handle.write("Context manager demo")
```

```python
# Advanced topic examples: collections, itertools, functools, regex, datetime, logging, typing
from collections import defaultdict, Counter, deque, namedtuple
import itertools as it
from functools import lru_cache, partial, wraps
import logging
from logging.handlers import RotatingFileHandler
import re
from datetime import datetime, timezone, timedelta
from pathlib import Path
from dataclasses import dataclass
from typing import Generic, TypeVar, Optional

# Collections
sales = defaultdict(int)
sales['apple'] += 2
sales['banana'] += 1
print('defaultdict:', dict(sales))

counts = Counter(['apple', 'apple', 'banana'])
print('Counter:', counts)

queue = deque(['a', 'b'])
queue.append('c')
queue.popleft()
print('deque:', list(queue))

Point = namedtuple('Point', ['x', 'y'])
print('namedtuple:', Point(3, 4))

# itertools
print('permutations:', list(it.permutations('AB', 2)))
print('combinations:', list(it.combinations([1, 2, 3], 2)))
print('product:', list(it.product([1, 2], ['x', 'y'])))

students = [('A', 90), ('A', 85), ('B', 80), ('B', 78)]
print('groupby:')
for key, group in it.groupby(students, key=lambda item: item[0]):
    print(key, list(group))

# functools
@lru_cache(maxsize=None)
def fib(n):
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)

print('lru_cache fib(8):', fib(8))

add_five = partial(lambda a, b: a + b, 5)
print('partial:', add_five(3))

def decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print('calling', func.__name__)
        return func(*args, **kwargs)
    return wrapper

@decorator
def greet(name):
    return f'Hello {name}'

print('wraps:', greet('Asha'))

# Regular expressions
text = 'Email: alice@example.com and bob@example.org'
pattern = re.compile(r'\b[a-z0-9._%+-]+@[a-z0-9.-]+\.[a-z]{2,}\b')
print('regex matches:', pattern.findall(text))

# Datetime
now = datetime.now(timezone.utc)
then = now + timedelta(days=3, hours=2)
print('datetime:', now)
print('timedelta:', then)

# Logging
log_path = Path('BasicPython/Level_1_Programs/output/advanced.log')
logger = logging.getLogger('advanced_demo')
logger.setLevel(logging.INFO)
logger.handlers.clear()
handler = RotatingFileHandler(log_path, maxBytes=200, backupCount=2)
logger.addHandler(handler)
logger.info('This is a log message')
print('logging file:', log_path.read_text(encoding='utf-8').strip())

# Typing and dataclasses
T = TypeVar('T')

class Box(Generic[T]):
    def __init__(self, value: T):
        self.value = value

@dataclass
class User:
    name: str
    age: int

print('box value:', Box(5).value)
print('dataclass:', User('Asha', 21))

try:
    import pydantic
    print('pydantic version:', pydantic.__version__)
except ImportError:
    print('pydantic not installed; install it with pip install pydantic')
```
