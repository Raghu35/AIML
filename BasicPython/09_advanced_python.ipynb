# 09. Advanced Python Concepts

This notebook covers decorators, generators, context managers, and more.

What this topic is:
These advanced tools allow you to write more flexible and elegant Python code.

Where we use it:
They are used in frameworks, libraries, and production-grade applications.

Why we use it:
They help reduce duplication and make complex behavior easier to manage.

```python
# Decorator example
def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper

@uppercase_decorator
def greet():
    return 'hello world'

print(greet())
```

```python
# Generator example
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for value in countdown(3):
    print(value)
```

```python
# Context manager example
from contextlib import contextmanager

@contextmanager
def file_manager(filename):
    f = open(filename, 'w', encoding='utf-8')
    try:
        yield f
    finally:
        f.close()

with file_manager('advanced_example.txt') as f:
    f.write('Advanced Python')
```
