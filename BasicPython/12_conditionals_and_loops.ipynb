# Conditionals and Loops

This notebook demonstrates if/else statements and loops in Python.

What this topic is:
Conditionals let programs make decisions, and loops repeat actions efficiently.

Where we use it:
They are used in validation, iteration, automation, and processing repeated tasks.

Why we use it:
They are essential for controlling program behavior and reducing repetitive code.

```python
# Set a mark value
marks = 85
```

```python
# Use if/elif/else
if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
else:
    grade = "C"

print("Grade:", grade)
```

```python
# For loop example
print("For loop:")
for i in range(1, 6):
    print(i)
```

```python
# While loop example
print("While loop:")
count = 1
while count <= 5:
    print(count)
    count += 1
```
