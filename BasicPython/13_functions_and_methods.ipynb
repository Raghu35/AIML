# Functions and Methods

This notebook shows how to define functions and use built-in methods.

What this topic is:
Functions group reusable logic, and methods are functions attached to objects.

Where we use it:
They appear in almost every Python program for reusable behavior and object interaction.

Why we use it:
They make code organized, reusable, and easier to understand.

```python
# Define a function
def add_numbers(a, b):
    return a + b

print("Sum:", add_numbers(10, 20))
```

```python
# Use built-in string methods
text = "python"
print("Uppercase:", text.upper())
print("Replaced:", text.replace("p", "J"))
print("Length:", len(text))
```
