# 03. Data Structures

This notebook covers lists, tuples, sets, and dictionaries.

What this topic is:
Data structures help us store and organize data in useful ways for different kinds of tasks.

Where we use it:
They are used in data processing, application state handling, and organizing large amounts of information.

Why we use it:
They make code more efficient and easier to work with.

```python
# A list stores multiple items in order
numbers = [1, 2, 3, 4]
numbers.append(5)
print(numbers)
```

```python
# A tuple is immutable, while a set keeps unique values
point = (10, 20)
unique_numbers = {1, 2, 2, 3}
print(point)
print(unique_numbers)
```

```python
# A dictionary stores data as key-value pairs
student = {"name": "Bob", "age": 22}
print(student["name"])
print(student.get("age"))
```
