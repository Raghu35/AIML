# 02. Control Flow and Functions

This notebook introduces conditionals, loops, and functions.

What this topic is:
Control flow decides the order of execution, while functions let us reuse logic in a structured way.

Where we use it:
These are used in decision-making logic, repetitive tasks, and reusable program components.

Why we use it:
They make programs smarter, shorter, and easier to maintain.

```python
# Use conditional logic to choose a grade
marks = 85
if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
else:
    grade = "C"
print("Grade:", grade)
```

```python
# A for loop repeats code a fixed number of times
for i in range(1, 6):
    print(i)
```

```python
# Functions let us group reusable logic in one place
def add_numbers(a, b):
    return a + b

print(add_numbers(10, 20))
```
