# Dictionary and Tuple Examples

This notebook demonstrates tuple immutability and dictionary key-value pairs.

What this topic is:
Dictionaries store data as key-value pairs, while tuples hold fixed collections of values.

Where we use it:
They are used in configuration data, records, mappings, and structured information.

Why we use it:
They provide simple and reliable ways to organize related data.

```python
# Tuple example
coordinates = (10, 20)
print("Tuple:", coordinates)
print("First value:", coordinates[0])
```

```python
# Dictionary example
student = {
    "name": "Bob",
    "age": 22,
    "course": "Python"
}

print("Dictionary:", student)
print("Name:", student["name"])
print("Course:", student.get("course"))
```
