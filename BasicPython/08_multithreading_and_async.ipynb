# 08. Multithreading and Async Programming

This notebook introduces threading and async features in Python.

What this topic is:
Multithreading and asyncio help Python programs handle multiple tasks at the same time.

Where we use it:
They are used in web servers, APIs, and input/output-heavy applications.

Why we use it:
They improve efficiency when programs spend time waiting on external resources.

```python
# Threading example
import threading
import time

def worker():
    time.sleep(1)
    print('Thread finished')

t = threading.Thread(target=worker)
t.start()
t.join()
```

```python
# Async example
import asyncio

async def main():
    print('Hello')
    await asyncio.sleep(0.1)
    print('World')

asyncio.run(main())
```
