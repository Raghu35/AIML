# 08. Multithreading and Async Programming

This notebook introduces threading and async features in Python.

**What it is:** Multithreading runs multiple threads in one process for I/O-bound work; AsyncIO uses async/await for cooperative concurrency without blocking.

**Where we use it:** Web servers, concurrent downloads, APIs, web scraping, and applications waiting on network or disk I/O.

**Why we use it:** They improve throughput when programs spend time waiting on external resources instead of computing.

> See [Module3/01_multithreading.ipynb](../Module3/01_multithreading.ipynb), [Module3/02_multiprocessing.ipynb](../Module3/02_multiprocessing.ipynb), and [Module3/03_asyncio.ipynb](../Module3/03_asyncio.ipynb).

```python
# Threading example
import threading
import time

def worker():
    time.sleep(1)
    print('Thread finished')

t = threading.Thread(target=worker)
t.start()
t.join()
```

```python
# Async example
import asyncio

async def main():
    print('Hello')
    await asyncio.sleep(0.1)
    print('World')

asyncio.run(main())
```

```python
# Multiprocessing — separate processes for CPU-bound parallelism (bypasses the GIL)
# See Module3/02_multiprocessing.ipynb for full examples.
from concurrent.futures import ProcessPoolExecutor

def square(n):
    return n * n

if __name__ == "__main__":
    with ProcessPoolExecutor(max_workers=2) as pool:
        results = list(pool.map(square, [1, 2, 3, 4, 5]))
    print("Squared:", results)
```
